```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
data = pd.read_csv("C:/Users/yasht/OneDrive/Documents/Student_performance_data 2.csv")
data
```

```python
data.isnull().sum()
```

```python
df=pd.DataFrame(data)
df
```

```python
gb=df.groupby("ParentalEducation").agg({"GPA":'mean'})
gb
```

```python
plt.figure(figsize=(5,5))
snb.heatmap(gb,annot=True)
plt.title("Results of students on the basis of parental education")
#parent education's impact on children .
```

```python
gm=df.groupby("Gender").agg({"GPA":'mean'})
gm
```

```python
snb.heatmap(gm,annot=True)
plt.title("Results on the basis of gender")
#results of both male and female on average we cannot find any difference 
```

```python
snb.boxplot(data=df,x="GPA")
```

```python
print(df["GradeClass"].unique())
```

```python
grade0=df.loc[(df["GradeClass"]==0)].count()
grade1=df.loc[(df['GradeClass']==1)].count()
grade2=df.loc[(df['GradeClass']==2)].count()
grade3=df.loc[(df['GradeClass']==3)].count()
grade4=df.loc[(df['GradeClass']==4)].count()



```

```python
#IT show percentage vise students in each grade
l=["grade0","grade1","grade2","grade3","grade4"]
mlist = [grade0["GradeClass"].sum(), grade1["GradeClass"].sum(), 
         grade2["GradeClass"].sum(), grade3["GradeClass"].sum(), 
         grade4["GradeClass"].sum()]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of students grade vise")
plt.show()
```
